```python
transaktioner = []          # En lista som loggar transaktioner
transaktioner.append(500)   # Adderar startbeloppet på 500 kr

def saldo():
    saldo = 500     # Ditt saldo börjar på 500 kr
    for t in transaktioner:
        saldo += t
    return saldo

def validate_int(output, error_mess):
    while True:
        try:
            value = int(input(output))
            break
        except:
            print(error_mess)
    return value    

def skriv_transaktion():
    line = 0
    summa = 0
    output = ("\n<=>   Transaktioner   <=>"
              "\n{:>3} {:>12} {:>12}"
              "\n≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖").format("Nr", "Händelse", "Saldo")
    for t in transaktioner:
        line += 1
        summa += t
        output += ("\n{:>2}. {:>9} kr {:>9} kr".format(line, t, summa))
    
    return output



# PROGRAMLOOP
while True:     

    meny = ("\n≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖\n"
            "\n           ▶ B͏a͏n͏k͏ o͏f͏ J͏a͏p͏a͏n͏ ◀"
            "\n\n≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖"
            "\n\n<=> 1. Redovisa Saldo             <=>"
            "\n<=> 2. Insättning                 <=>"
            "\n<=> 3. Uttag                      <=>"
            "\n<=> 0. Avsluta/Stoppa programmet  <=>"
            "\n\n≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖≖"
            "\n<=> Gör ditt val:                 <=>\n\n"
            "").format(saldo())


    välj = validate_int(meny, "Felaktig Inmatning!")

    if välj == 0:               # Avslutar/Stoppar programmet
        break

    elif välj == 1:             # Redovisar Transaktioner
        print(skriv_transaktion())



    elif välj == 2:
        print("<=>   Insättning   <=>")
        insättning = validate_int(input("<=>  Ange Belopp:  <=>", "⊘ Felaktig Inmatning!"))
        if insättning > 0:
            saldo += insättning
            transaktioner.append(insättning)
        else:
            print("⊘ Insättningen måste vara större än 0! ⊘")

    elif välj == 3:
        print("<=>     Uttag     <=>")
        uttag = validate_int(input("<=>  Ange Belopp:  <=>"))
        if uttag <= saldo() and uttag >= 0:
            saldo -= uttag
            transaktioner.append(-uttag)
        elif uttag < 0:
            print("⊘ Uttaget måste vara större än 0! ⊘")
        else:
            print("⊘ Beloppet får inte vara större än saldo! ⊘")


    else:
        print("<=>  Felaktigt val!  <=>")

print("\nTack för ditt besök, välkommen åter!")
```
